```python
import pandas as pd
```

```python
df = pd.read_csv('jogadores_base.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
extra_attributes = df["Extra"].str.split(expand=True)

extra_attributes.columns = ["Extra_" + str(i) for i in range(extra_attributes.shape[1])]
df.drop(columns=['Extra'], inplace=True)

df = pd.concat([df, extra_attributes], axis=1)
```

```python
unique_caracteristicas = df['Caracteristica 1'].unique()
unique_caracteristicas
```

```python
caracteristicas = {
    0: "Colocação",
    1: "Defesa de Pênalti",
    2: "Reflexo",
    3: "Saída de Gol",
    4: "Armação",
    5: "Cabeceio",
    6: "Cruzamento",
    7: "Desarme",
    8: "Drible",
    9: "Finalização",
    10: "Marcação",
    11: "Passe",
    12: "Resistência",
    13: "Velocidade"
}

df['Caracteristica 1'] = df['Caracteristica 1'].map(caracteristicas)
df['Caracteristica 2'] = df['Caracteristica 2'].map(caracteristicas)
```

```python
posicoes = {
    0: "Goleiro",
    1: "Lateral",
    2: "Zagueiro",
    3: "Meio Campo",
    4: "Atacante"
}
df['Posicao'] = df['Posicao'].map(posicoes)
```

```python
df.drop(columns=['Extra_0'], inplace=True)
df.drop(columns=['Extra_2'], inplace=True)
df.drop(columns=['Extra_3'], inplace=True)
df
```

```python
df['Extra_1'] = df['Extra_1'].str.extract(r'=(\d+)')
df
```

```python
resultado = df[df['Extra_1'] == '10']

goleiros = resultado[df['Posicao'] == 'Goleiro']
laterais = resultado[df['Posicao'] == 'Lateral']
zagueiros = resultado[df['Posicao'] == 'Zagueiro']
meio_de_campo = resultado[df['Posicao'] == 'Meio Campo']
atacantes = resultado[df['Posicao'] == 'Atacante']
```

```python
goleiros = goleiros[(goleiros['Posicao'] == 'Goleiro') & (goleiros['Idade'] >= 17)].sample(n=2)

laterais_esquerdo = laterais[(laterais['Lado'] == 'Esquerdo') & (laterais['Idade'] == 20)].sample(n=2)
laterais_direito = laterais[(laterais['Lado'] == 'Direito') & (laterais['Idade'] >= 18)].sample(n=2)


zagueiros_esquerdo = zagueiros[(zagueiros['Lado'] == 'Esquerdo') & (zagueiros['Idade'] == 20)].sample(n=2)
zagueiros_direito = zagueiros[(zagueiros['Lado'] == 'Direito') & (zagueiros['Idade'] >= 18)].sample(n=2)

volante = meio_de_campo[(meio_de_campo['Caracteristica 1'] == 'Desarme') & (meio_de_campo['Idade'] >= 19)]
meia = meio_de_campo[(meio_de_campo['Caracteristica 1'] == 'Armação') & (meio_de_campo['Idade'] >= 19)]

volante_esquerdo = volante[volante['Lado'] == 'Esquerdo'].sample(n=2)
volante_direito = volante[volante['Lado'] == 'Direito'].sample(n=2)

meia_esquerdo = meia[meia['Lado'] == 'Esquerdo'].sample(n=2)
meia_direita = meia[meia['Lado'] == 'Direito'].sample(n=2)

atacante_esquerdo = atacantes[(atacantes['Idade'] >= 19) & (atacantes['Caracteristica 1'] == 'Finalização') & (atacantes['Lado'] == 'Esquerdo')].sample(n=2)
atacante_direito = atacantes[(atacantes['Idade'] >= 19) & (atacantes['Caracteristica 1'] == 'Finalização') & (atacantes['Lado'] == 'Direito')].sample(n=2)

pontas_esquerdo = atacantes[(atacantes['Idade'] >= 19) & (atacantes['Caracteristica 1'] == 'Drible') & (atacantes['Lado'] == 'Esquerdo')].sample(n=2)
pontas_direito = atacantes[(atacantes['Idade'] >= 19) & (atacantes['Caracteristica 1'] == 'Drible') & (atacantes['Lado'] == 'Direito')].sample(n=2)

x = pd.concat([goleiros, laterais_esquerdo, laterais_direito,zagueiros_esquerdo, zagueiros_direito, volante_direito, volante_esquerdo, meia_esquerdo,meia_direita, atacante_esquerdo, atacante_direito, pontas_esquerdo,pontas_direito])
x
```
